```python
df=df.fillna(df.mean())
```

```python
df
```

```python
Y=df['hospital_death'].values
Y=Y.astype(int)
type(Y)
```

```python
X=df.drop(labels=['hospital_death'],axis=1)
type(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train.size
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,random_state=0, criterion='entropy')
model.fit(X_train,Y_train)
```

```python
predict_test=model.predict(X_test)
predict_test.size
```

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,precision_score,recall_score
print("Accuracy=",accuracy_score(Y_test,predict_test) * 100)
#plot_confusion_matrix(Y_test,predict_test)
#sensitivity = cm[0,0]/(cm[0,0]+cm[0,1])
#specificity = cm[1,1]/(cm[1,0]+cm[1,1])
#print('Sensitivity= ' ,sensitivity)
#print('Specificity= ',specificity)
#print('Precision= ',precision_score(Y_test,predict_test))
#print('Recall= ',recall_score(Y_test,predict_test))
```

```python
feature_list=list(X.columns)
feature_imp=pd.Series(model.feature_importances_,index=feature_list).sort_values(ascending=False)
feature_imp.plot(kind='barh', figsize=(10,10))
plt.xlabel('Precentage')
plt.ylabel('Parameters')
```

```python
feature_list=list(X.columns)
for feature in zip(feature_list,model.feature_importances_):
    print(feature)
```

```python
from sklearn.feature_selection import SelectFromModel
par=SelectFromModel(model,threshold=0.012)
```

```python
par.fit(X_train,Y_train)
```

```python
for feature_list_index in par.get_support(indices=True):
    f_list=feature_list[feature_list_index]
    print(f_list)
```

```python
X_important_train = par.transform(X_train)
X_important_test = par.transform(X_test)
```

```python
clf_important = RandomForestClassifier(n_estimators=100, random_state=0, criterion='entropy')
clf_important.fit(X_important_train, Y_train)
```

```python
y_predict=clf_important.predict(X_important_test)
accuracy_score(Y_test,y_predict)
```

```python
for feature_important in zip(f_list,clf_important.feature_importances_):
    print(feature_important)
```
